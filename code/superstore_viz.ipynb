```python
loss_df = dx[dx['Profit']<0]
loss_df#negative profit data (loss data)
```

```python
loss_df.shape
```

```python
loss_df.describe()
```

```python
import numpy as np
```

```python
total_loss = np.negative(loss_df['Profit'].sum())
print('Total loss = %.2f' %total_loss)
```

```python
loss_df.groupby(by = 'Segment').sum()
```

```python
loss_df.groupby(by = 'Category').sum()
```

```python
loss_df.groupby(by = 'SubCategory').sum()
```

```python
loss_df['SubCategory'].value_counts()
```

```python
loss_df.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)#top 10 cities with highest loss
```

```python
loss_df.groupby(by='Category').mean().sort_values('Sales',ascending=True)
```

```python
loss_df.groupby(by='Category').sum().sort_values('Sales',ascending=True)
```

```python
dx.groupby(['State']).sum()['Sales'].nsmallest(10)# last 10 states with least sales.
```

```python
loss_df.groupby(by='Segment').sum().sort_values('Sales',ascending=True)
```

```python
dx.sort_values(['Sales'],ascending=True).groupby('Segment').sum()
```

```python
dx.groupby(by='Region').sum()
```

# Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(15,3)#set figure size
```

```python
plt.bar(loss_df['SubCategory'],loss_df['Sales']);
plt.rcParams.update({'font.size':10});
plt.xlabel('Sub Category');
plt.ylabel('Sales');
```

```python
plt.rcParams['figure.figsize']=(28,8)
plt.bar(dx['SubCategory'],dx['Sales']);
plt.rcParams.update({'font.size':14});
plt.xlabel('Sub Category');
plt.ylabel('Sales');
```

```python
plt.rcParams['figure.figsize']=(28,8)
plt.bar(dx['SubCategory'],dx['Discount']);
plt.rcParams.update({'font.size':14});
plt.xlabel('Sub_Category');
plt.ylabel('Discount');
```

```python
plt.rcParams['figure.figsize']=(10,8)
plt.bar(dx['ShipMode'],dx['Sales']);
plt.rcParams.update({'font.size':14});
plt.xlabel('Ship Mode');
plt.ylabel('Sales');
```
